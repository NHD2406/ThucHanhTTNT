# CÂU 1 (2 ĐIỂM) – Quay Lui: Sinh Hoán Vị & Derangement

Bài toán **derangement** (hoán vị xáo trộn): Cho mảng gốc `goc = [1,2,3,4]`, một **derangement** là hoán vị mà **không phần tử nào đứng ở vị trí ban đầu**.

> *Ví dụ với gốc `[1,2,3,4]`:*
> - `[2,1,4,3]` ✓ là derangement (không vị trí nào trùng)
> - `[2,1,3,4]` ✗ KHÔNG phải (vì `3` ở vị trí 2 và `4` ở vị trí 3 — trùng gốc)
> - Số derangement của `[1,2,3,4]` = **9** (kí hiệu `D₄ = 9`).

**b) (0.3đ)** Chạy `hoan_vi(arr=[1,2,3,4])` và đếm tổng số hoán vị in ra. Verify = `4! = 24`.

**c) (0.6đ)** Gọi `derangement([1,2,3,4], [1,2,3,4], 0, [])`. In tất cả derangement và đếm. Kiểm tra = **9**. In thêm số lần gọi đệ quy `dem_d`.

**d) (0.4đ) – So sánh hiệu năng PRUNING:** Cài đặt phiên bản **KHÔNG pruning** `derangement_no_prune(arr, goc, start, ket_qua)` — sinh tất cả 24 hoán vị rồi lọc derangement ở cuối. So sánh số lần gọi đệ quy với phiên bản có pruning. Giải thích bằng comment vì sao pruning hiệu quả hơn.

````python
# CAU 1a – Nhin anh phia tren, go lai toan bo code va dien 7 cho trong
# Khong duoc copy tu nguon khac
# ============================================================
dem_goi = 0
def hoan_vi(arr, start=0):
    """Sinh tat ca hoan vi cua arr (in-place swap)."""
    global dem_goi
    dem_goi += 1

    # TRONG 1: Dieu kien dung
    if start == len(arr):
        print(arr.copy())
        return

    # TRONG 2: Vong lap duyet tu vi tri start
    for i in range(start, len(arr)):
        # TRONG 3: Trao doi phan tu
        arr[start], arr[i] = arr[i], arr[start]

        # TRONG 4: Goi de quy cho vi tri ke tiep
        hoan_vi(arr, start + 1)

        # QUAY LUI: undo swap (tra lai trang thai cu)
        # TRONG 5: Khoi phuc mang ban dau
        arr[start], arr[i] = arr[i], arr[start]

# --- PHẦN 2: DERANGEMENT (Hoán vị không có phần tử nào ở vị trí gốc) ---
# Vi du goc=[1,2,3,4]: [2,1,4,3] OK | [2,1,3,4] KO (3 va 4 con trung goc)

dem_d = 0

def derangement(arr, goc, start, ket_qua):
    """Sinh tat ca derangement voi PRUNING SOM (cat nhanh som)."""
    global dem_d
    dem_d += 1

    if start == len(arr):
        ket_qua.append(arr.copy())
        return

    for i in range(start, len(arr)):
        # PRUNING: phan tu arr[i] se trung vi tri goc -> bo qua nhanh nay
        # TRONG 6: So sanh phan tu sap swap voi vi tri goc tuong ung
        if arr[i] == goc[start]:
            continue

        arr[start], arr[i] = arr[i], arr[start]

        # TRONG 7: De quy tiep tuc
        derangement(arr, goc, start + 1, ket_qua)

        # Quay lui
        arr[start], arr[i] = arr[i], arr[start]
````

````python
# CAU 1b – Run hoan_vi va dem so hoan vi
# ============================================================

# Đảm bảo bạn đã định nghĩa hàm hoan_vi ở phía trên
dem_goi = 0

def hoan_vi(arr, start=0):
    global dem_goi
    dem_goi += 1

    # Điều kiện dừng: Khi start bằng độ dài mảng, ta có 1 hoán vị hoàn chỉnh
    if start == len(arr):
        print(arr)
        return

    for i in range(start, len(arr)):
        # Tráo đổi (Swap)
        arr[start], arr[i] = arr[i], arr[start]

        # Đệ quy cho vị trí tiếp theo
        hoan_vi(arr, start + 1)

        # Quay lui (Backtrack) để trả lại trạng thái mảng ban đầu
        arr[start], arr[i] = arr[i], arr[start]

# --- THỰC THI CÂU B ---
print("Tat ca hoan vi cua [1,2,3,4]:")
print("-" * 30)

# Reset biến đếm trước khi chạy
dem_goi = 0
hoan_vi([1, 2, 3, 4])

print("-" * 30)
print(f"So lan goi de quy hoan_vi: {dem_goi}")
print(f"Ket qua kiem tra: 4! = {4*3*2*1}")
print("Giai thich: Moi dong in ra la mot hoan vi. Co tong cong 24 dong.")
````

````python
# CAU 1d – So sanh PRUNING vs KHONG PRUNING
# ============================================================

dem_no = 0

def derangement_no_prune(arr, goc, start, ket_qua):
    """Sinh TAT CA hoan vi, sau do loc derangement (KHONG pruning)."""
    global dem_no
    dem_no += 1

    # Base case: Khi đã sinh xong một hoán vị hoàn chỉnh
    if start == len(arr):
        # Kiểm tra điều kiện derangement: không phần tử nào trùng vị trí gốc
        is_derangement = True
        for i in range(len(arr)):
            if arr[i] == goc[i]:
                is_derangement = False
                break

        if is_derangement:
            ket_qua.append(arr.copy())
        return

    # Sinh hoán vị thông thường (không có điều kiện lọc ở giữa chừng)
    for i in range(start, len(arr)):
        arr[start], arr[i] = arr[i], arr[start]
        derangement_no_prune(arr, goc, start + 1, ket_qua)
        arr[start], arr[i] = arr[i], arr[start]

# --- CHẠY SO SÁNH ---

# 1. Chạy có Pruning (sử dụng lại kết quả dem_d từ câu c hoặc chạy lại)
dem_d = 0
ket_qua_prune = []
derangement([1, 2, 3, 4], [1, 2, 3, 4], 0, ket_qua_prune)

# 2. Chạy không Pruning
dem_no = 0
ket_qua_no = []
derangement_no_prune([1, 2, 3, 4], [1, 2, 3, 4], 0, ket_qua_no)

print("=" * 55)
print("SO SANH: PRUNING vs KHONG PRUNING")
print("=" * 55)
print(f"Co PRUNING     | so derangement: {len(ket_qua_prune)} | so de quy: {dem_d}")
print(f"KHONG PRUNING  | so derangement: {len(ket_qua_no)}     | so de quy: {dem_no}")

if dem_no > 0:
    tiet_kiem = (1 - dem_d/dem_no) * 100
    print(f"Pruning tiet kiem: {tiet_kiem:.1f}% so lan goi")

# --- GIẢI THÍCH (COMMENT) ---
"""
GIẢI THÍCH HIỆU QUẢ CỦA PRUNING:

- KHÔNG pruning: Thuật toán phải duyệt đến tận cùng của cây đệ quy (tất cả 24 hoán vị)
  mới bắt đầu kiểm tra điều kiện. Nghĩa là ngay cả khi số đầu tiên đã sai (ví dụ: [1, ...]),
  nó vẫn cố gắng sinh tiếp các số phía sau, gây lãng phí rất nhiều lần gọi đệ quy (65 lần).

- CÓ pruning: Thuật toán kiểm tra điều kiện ngay tại mỗi bước 'thử'. Nếu phát hiện
  phần tử sắp tráo đổi gây trùng vị trí gốc (arr[i] == goc[start]), nó sẽ 'ngắt'
  nhánh đó ngay lập tức (continue). Nhờ việc phát hiện sai lầm sớm ở các tầng trên,
  toàn bộ các nhánh con phía dưới không cần phải sinh ra, giúp giảm số lần gọi
  đệ quy xuống đáng kể (chỉ còn 25 lần).
"""
````

**Câu 2a – What-if (0.25đ bonus):**

> Giả sử đổi `h(D) = 8` (thay vì 4). Lúc này h(D) = 8 > h*(D) = 5 → **không admissible** nữa.  
> A* có còn tìm được đường đi **tối ưu** không? Vì sao?

*Heuristic h(D) = 8 là một giá trị ước lượng quá cao so với chi phí thực tế từ D đến G (h*(D) = 4). Điều này khiến thuật toán A* đánh giá đường đi qua nút D trở nên kém hấp dẫn hơn so với thực tế. Khi A* gặp nút D, tổng chi phí dự kiến f(D) = g(D) + h(D) sẽ rất lớn, làm cho các đường đi khác (có thể dài hơn nhưng có heuristic không bị ước lượng quá cao) được ưu tiên khám phá trước. Kết quả là, A* có thể tìm thấy một đường đi đến G có chi phí cao hơn (ví dụ: S -> B -> E -> G với tổng chi phí 8) và kết luận đó là đường đi tốt nhất, bỏ lỡ đường đi tối ưu thực sự (S -> A -> D -> E -> G với tổng chi phí 7).*

````python
# CAU 2b – Cai dat A*
# Tu go lai graph va heuristic tu do thi roi cai dat astar()
# ============================================================
import heapq

# TODO: Tu go lai graph tu do thi (7 nut: S, A, B, C, D, E, G)
graph = {
    'S': {'A': 2, 'B': 5},
    'A': {'C': 4, 'D':1},
    'B': {'D': 2,'E':4},
    'C':{'A':4, 'G':3},
    'D':{'A':1, 'B':2,'G':5},
    'E':{'G':2,'B':4},
    'G':{}
}

# TODO: Tu go lai heuristic tu do thi
heuristic = {
    'S': 7,
    'A': 4,
    'B': 5,
    'C': 3,
    'D': 4,
    'E': 2,
    'G':0

}

def astar(graph, heuristic, start, goal):
    # TODO
    pass
    open_list = []  # priority queue: (f_cost, g_cost, node, path_list)
    # f_cost = g_cost + h_cost. Cho nút bắt đầu, g_cost là 0.
    heapq.heappush(open_list, (heuristic[start], 0, start, [start]))

    # g_costs: lưu trữ chi phí g_cost thấp nhất hiện tại để đến mỗi nút
    g_costs = {node: float('inf') for node in graph}
    g_costs[start] = 0

    closed_list = set()  # Các nút đã được mở rộng
    node_expanded_count = 0
    step_count = 0

    while open_list:
        f_current, g_current, current_node, path_current = heapq.heappop(open_list)

        # Nếu đã tìm thấy đường tốt hơn hoặc đã xử lý nút này với đường dẫn này, bỏ qua
        # Điều này xử lý các trường hợp khi một nút được đẩy nhiều lần với các g_cost khác nhau.
        if g_current > g_costs[current_node]:
            continue

        node_expanded_count += 1
        step_count += 1

        # Nếu đạt được mục tiêu
        if current_node == goal:
            h_current = heuristic[current_node]
            f_val_to_print = g_current + h_current

            # Xây dựng chuỗi danh sách OPEN để in (chỉ các mục hợp lệ còn trong heap)
            temp_open_for_print = {}
            for f_val_heap, g_val_heap, node_heap, _ in open_list:
                if g_val_heap == g_costs[node_heap]:  # Đảm bảo không phải mục lỗi thời
                    if node_heap not in temp_open_for_print or f_val_heap < temp_open_for_print[node_heap]:
                        temp_open_for_print[node_heap] = f_val_heap

            # Sắp xếp các mục trong chuỗi OPEN theo f-cost, sau đó theo tên nút (bảng chữ cái)
            sorted_open = sorted(temp_open_for_print.items(), key=lambda item: (item[1], item[0]))
            open_str = ", ".join([f"{node}:{f}" for node, f in sorted_open])

            # Sắp xếp closed_list để đầu ra nhất quán
            closed_str = ", ".join(sorted(list(closed_list)))
            print(f"Buoc {step_count} | Xet: {current_node} | g={g_current} h={h_current} f={f_val_to_print} | OPEN: [{open_str}] | CLOSE: {{{closed_str}}}")
            return path_current, g_current, node_expanded_count

        # Khám phá các nút lân cận
        for neighbor, cost in graph[current_node].items():
            tentative_g_cost = g_current + cost

            if tentative_g_cost < g_costs[neighbor]:  # Tìm thấy đường tốt hơn đến nút lân cận
                g_costs[neighbor] = tentative_g_cost
                f_cost = tentative_g_cost + heuristic[neighbor]
                new_path = path_current + [neighbor]
                heapq.heappush(open_list, (f_cost, tentative_g_cost, neighbor, new_path))

        # In trace cho bước này SAU KHI xử lý tất cả các nút lân cận
        h_current = heuristic[current_node]
        f_val_to_print = g_current + h_current

        # Xây dựng chuỗi danh sách OPEN để in
        temp_open_for_print = {}
        for f_val_heap, g_val_heap, node_heap, _ in open_list:
            if g_val_heap == g_costs[node_heap]:  # Đảm bảo nó không phải là một mục lỗi thời
                if node_heap not in temp_open_for_print or f_val_heap < temp_open_for_print[node_heap]:
                    temp_open_for_print[node_heap] = f_val_heap

        # Sắp xếp các mục trong chuỗi OPEN theo f-cost, sau đó theo tên nút (bảng chữ cái)
        sorted_open = sorted(temp_open_for_print.items(), key=lambda item: (item[1], item[0]))
        open_str = ", ".join([f"{node}:{f}" for node, f in sorted_open])

        # Sắp xếp closed_list để đầu ra nhất quán
        closed_str = ", ".join(sorted(list(closed_list)))

        print(f"Buoc {step_count} | Xet: {current_node} | g={g_current} h={h_current} f={f_val_to_print} | OPEN: [{open_str}] | CLOSE: {{{closed_str}}}")

    return None, float('inf'), node_expanded_count

print("=" * 55)
print("A* – Tim duong tu S den G")
print("=" * 55)
path, cost, n_exp = astar(graph, heuristic, 'S', 'G')
if path:
    print(f"\nDuong di: {' -> '.join(path)}")
    print(f"Tong chi phi: {cost}")
    print(f"So node expand: {n_exp}")
````

**Câu 2c – Kiểm tra admissible:**

*(Tính h*(n) = chi phí thực tế ngắn nhất từ n đến G)*

- h*(S) = 6  →  h(S)=7  →  h(S) ≤ h*(S)? : **Không**
- h*(A) = 4  →  h(A)=4  →  h(A) ≤ h*(A)? : **Có**
- h*(D) = 3  →  h(D)=4  →  h(D) ≤ h*(D)? : **Không**
- h*(B) = 5  →  h(B)=5  →  h(B) ≤ h*(B)? : **Có**

**Kết luận:** Heuristic này **Không admissible** vì h(S) > h*(S) và h(D) > h*(D). Một heuristic admissible phải luôn thỏa mãn h(n) ≤ h*(n) cho tất cả các nút.

````python
# CAU 2d – Cai dat UCS (Uniform Cost Search) va so sanh voi A*
# UCS = A* voi h(n) = 0 cho moi node (tuong duong Dijkstra)
# ============================================================
import heapq

def ucs(graph, start, goal):
    """
    Uniform Cost Search.
    Chi dung g(n) - khong dung heuristic.
    Tra ve (duong_di, tong_chi_phi, so_node_expand).
    """
    open_list = []  # priority queue: (g_cost, node, path_list)
    heapq.heappush(open_list, (0, start, [start]))

    g_costs = {node: float('inf') for node in graph}
    g_costs[start] = 0

    # closed_list is not strictly needed for correctness but useful for tracking expanded nodes
    # to prevent re-adding nodes to open_list with worse costs. However, in UCS, if a node is
    # popped, its g_cost is guaranteed to be the shortest path found so far (like Dijkstra).
    node_expanded_count = 0

    while open_list:
        g_current, current_node, path_current = heapq.heappop(open_list)

        # If we found a better path to current_node already, skip this one
        if g_current > g_costs[current_node]:
            continue

        node_expanded_count += 1

        # Goal check
        if current_node == goal:
            return path_current, g_current, node_expanded_count

        # Explore neighbors
        for neighbor, cost in graph[current_node].items():
            tentative_g_cost = g_current + cost

            if tentative_g_cost < g_costs[neighbor]:  # Found a shorter path to neighbor
                g_costs[neighbor] = tentative_g_cost
                new_path = path_current + [neighbor]
                heapq.heappush(open_list, (tentative_g_cost, neighbor, new_path))

    return None, float('inf'), node_expanded_count # Goal not reachable


# So sanh A* va UCS
print("Executing A*...")
path_a, cost_a, exp_a = astar(graph, heuristic, 'S', 'G')
print("Executing UCS...")
path_u, cost_u, exp_u = ucs(graph, 'S', 'G')

print("=" * 55)
print("SO SANH A* vs UCS")
print("=" * 55)
print(f"{'Thuat toan':<12} {'Duong di':<22} {'Cost':<6} {'#Expand':<10}")
print("-" * 55)
print(f"{'A*':<12} {' -> '.join(path_a):<22} {cost_a:<6} {exp_a:<10}")
print(f"{'UCS':<12} {' -> '.join(path_u):<22} {cost_u:<6} {exp_u:<10}")

# TODO (comment): Nhan xet
# - Hai thuat toan co cung tong chi phi khong?
#   Có, cả A* và UCS đều tìm ra đường đi tối ưu với tổng chi phí là 8. Điều này là do cả hai
#   đều là thuật toán tìm kiếm tối ưu trên đồ thị không có cạnh âm (UCS tương đương Dijkstra, A* là thuật toán cải tiến của Dijkstra).
# - Thuat toan nao expand it node hon? Vi sao?
#   A* expand ít node hơn (4 node) so với UCS (6 node). Lý do là A* sử dụng hàm heuristic
#   h(n) để ước lượng chi phí còn lại từ node hiện tại đến đích. Heuristic này giúp A*
#   'thông minh' hơn, tập trung tìm kiếm vào những hướng có khả năng dẫn đến đích nhanh nhất,
#   tránh khám phá những nhánh không hiệu quả. Ngược lại, UCS không có heuristic, chỉ mở rộng
#   node dựa trên chi phí thực tế từ điểm bắt đầu, nên nó khám phá rộng hơn.
# - Khi nao A* va UCS cho ket qua giong nhau?
#   - Về đường đi và chi phí tối ưu: Cả A* (với heuristic admissible) và UCS đều luôn tìm ra
#     đường đi tối ưu nếu có. Vậy nên kết quả về đường đi và chi phí thường sẽ giống nhau.
#   - Về số lượng node được mở rộng: A* và UCS sẽ mở rộng số node bằng nhau khi hàm heuristic h(n) = 0 cho mọi node.
#     Trong trường hợp đó, A* chính là thuật toán UCS.
````

**Câu 3b – Tính tay state #1 (lượt X):**

Bảng:
```
X | X | _
O | O | _
_ | _ | _
```

Ô trống: `(0,2), (1,2), (2,0), (2,1), (2,2)`

Phân tích từng nước đi của X:
- **X đi `(0,2)`** → hàng 0 = `X,X,X` → **X thắng ngay** → utility = `1`
- **X đi `(1,2)`** → chặn O (vì O đe dọa hàng 1). Game tiếp tục, kết quả = `0` (hòa) nếu chơi tối ưu.
- **X đi `(2,0)`, `(2,1)`, hoặc `(2,2)`** → KHÔNG chặn O → O sẽ đi `(1,2)` → hàng 1 = `O,O,O` → **O thắng** → utility = `-1`

X là MAX → chọn nước có `min_value` LỚN nhất.  
**Kết luận:** Minimax chọn ô `(0,2)`

**Câu 3c – Tính tay state #2 (lượt O):**

Bảng:
```
X | X | _
_ | O | _
_ | _ | _
```

X đang đe dọa thắng tại ô `(0,2)` (X có 2 quân liên tiếp ở hàng 0).

Nếu O **không chặn**, lượt sau X sẽ thắng → utility = +1 → tệ cho O.  
Vậy O (MIN) phải chọn ô để **chặn** X.

**Kết luận:** Minimax (O) chọn ô `(0,2)`

````python
import copy, math

X, O, EMPTY = 'X', 'O', None

# --- CÁC HÀM CƠ BẢN (ĐÃ CÓ TRONG ĐỀ BÀI) ---

def initial_state():
    return [[EMPTY]*3 for _ in range(3)]

def player(board):
    x_count = sum(row.count(X) for row in board)
    o_count = sum(row.count(O) for row in board)
    return X if x_count == o_count else O

def actions(board):
    return {(i, j) for i in range(3) for j in range(3)
        if board[i][j] == EMPTY}

def result(board, action):
    i, j = action
    if board[i][j] is not EMPTY: raise ValueError()
    new_board = copy.deepcopy(board)
    new_board[i][j] = player(board)
    return new_board

def winner(board):
    for p in [X, O]:
        for row in board:
            if all(c == p for c in row): return p
        for col in range(3):
            if all(board[r][col] == p for r in range(3)): return p
        if all(board[i][i] == p for i in range(3)): return p
        if all(board[i][2 - i] == p for i in range(3)): return p
    return None

def terminal(board):
    return winner(board) is not None or not actions(board)

def utility(board):
    w = winner(board)
    return 1 if w == X else (-1 if w == O else 0)

# --- THUẬT TOÁN MINIMAX ---

def max_value(state):
    if terminal(state): return utility(state)
    v = -math.inf
    for a in actions(state):
        v = max(v, min_value(result(state, a)))
    return v

def min_value(state):
    if terminal(state): return utility(state)
    v = math.inf
    for a in actions(state):
        v = min(v, max_value(result(state, a)))
    return v

def minimax(board):
    if terminal(board): return None
    if player(board) == X:
        return max(actions(board), key=lambda a: min_value(result(board, a)))
    else:
        return min(actions(board), key=lambda a: max_value(result(board, a)))

# --- HÀM IN BẢNG (CẦN THIẾT ĐỂ HIỂN THỊ) ---

def print_board(board):
    print("\n  0 1 2") # In chỉ số cột
    for i, row in enumerate(board):
        row_display = [c if c is not None else "-" for c in row]
        print(f"{i} {' '.join(row_display)}")
    print()

# --- VÒNG LẶP GAME (YÊU CẦU CHÍNH) ---

def play_game():
    """
    Vong lap game: Nguoi (X) vs AI (O).
    """
    board = initial_state()
    print("=" * 40)
    print("      TIC-TAC-TOE: NGƯỜI (X) VS AI (O)")
    print("  Nhập tọa độ hàng và cột từ 0 đến 2.")
    print("  Ví dụ: '1 2' (hàng 1, cột 2)")
    print("=" * 40)
    print_board(board)

    while not terminal(board):
        current = player(board)

        if current == X:
            # Lượt của Người
            while True:
                try:
                    move_input = input(f"Lượt {current}. Nhập 'hàng cột': ")
                    # Tách chuỗi và chuyển thành int
                    row, col = map(int, move_input.split())
                    action = (row, col)

                    # Kiểm tra tọa độ trong khoảng [0, 2] và ô phải trống
                    if 0 <= row < 3 and 0 <= col < 3:
                        if board[row][col] == EMPTY:
                            board = result(board, action)
                            break
                        else:
                            print("❌ Ô này đã được đánh rồi! Hãy chọn ô khác.")
                    else:
                        print("❌ Tọa độ ngoài bàn cờ (phải từ 0-2).")
                except (ValueError, IndexError):
                    print("❌ Sai định dạng! Vui lòng nhập 2 số cách nhau bởi khoảng trắng (vd: 1 2).")
        else:
            # Lượt của AI
            print(f"🤖 AI đang suy nghĩ (O)...")
            action = minimax(board)
            if action is not None:
                board = result(board, action)
                print(f"🤖 AI đã đánh vào ô: {action}")

        print_board(board)

    # Kết thúc game
    res_winner = winner(board)
    print("=" * 40)
    if res_winner == X:
        print("🏆 KẾT QUẢ: BẠN (X) THẮNG!")
    elif res_winner == O:
        print("💀 KẾT QUẢ: AI (O) THẮNG!")
    else:
        print("🤝 KẾT QUẢ: HÒA!")
    print("=" * 40)

# Chạy game
if __name__ == "__main__":
    play_game()
````

**Câu 3e – Tại sao Minimax chậm và đề xuất cải tiến:**

(Minimax chậm với bàn lớn hơn 3x3 chủ yếu là do sự bùng nổ tổ hợp (combinatorial explosion):

1.  **Branching Factor Lớn:** Khi kích thước bàn cờ tăng lên, số lượng nước đi hợp lệ ở mỗi trạng thái cũng tăng lên đáng kể, làm cho 'độ rộng' của cây tìm kiếm tăng theo cấp số nhân.
2.  **Độ Sâu Tìm Kiếm Lớn:** Để tìm ra nước đi tối ưu, Minimax cần duyệt sâu xuống các nước đi có thể xảy ra trong tương lai. Với branching factor lớn, ngay cả việc tìm kiếm vài nước đi sâu cũng tạo ra một số lượng trạng thái khổng lồ cần phải đánh giá.

**Đề xuất cải tiến:**
Một cải tiến cụ thể là sử dụng **Alpha-Beta Pruning (cắt tỉa Alpha-Beta)**. Kỹ thuật này giúp loại bỏ các nhánh của cây tìm kiếm mà không cần phải đánh giá, vì nó nhận ra rằng các nhánh đó sẽ không bao giờ dẫn đến nước đi tối ưu cho người chơi hiện tại. Nó giảm đáng kể số lượng node cần duyệt mà không làm mất đi tính chính xác của thuật toán Minimax.)
