### Hàm trộn 2 mảng một chiều

```python
"""Bài 13"""
def tron_mang_voi_tong(mang1, mang2):
    mang_ket_qua = []
    do_dai1 = len(mang1)
    do_dai2 = len(mang2)
    do_dai_min = min(do_dai1, do_dai2)
    for i in range(do_dai_min):
        mang_ket_qua.append(mang1[i] + mang2[i])
    if do_dai1 > do_dai_min:
        mang_ket_qua.extend(mang1[do_dai_min:])
    elif do_dai2 > do_dai_min:
        mang_ket_qua.extend(mang2[do_dai_min:])
    return mang_ket_qua
mang_a = [3, 9, 1, 4]
mang_b = [2, 7, 4, 3, 2, 8]
ket_qua_tron = tron_mang_voi_tong(mang_a, mang_b)
print(f"Mảng a: {mang_a}")
print(f"Mảng b: {mang_b}")
print(f"Mảng kết quả: {ket_qua_tron}\n")

mang_x = [1, 2, 3, 4, 5]
mang_y = [6, 7, 8, 9, 2]
ket_qua_tron_xy = tron_mang_voi_tong(mang_x, mang_y)
print(f"Mảng x: {mang_x}")
print(f"Mảng y: {mang_y}")
print(f"Mảng kết quả: {ket_qua_tron_xy}")
```

```python
"""Bài 14"""
import random

def tao_mang_ngau_nhien(m, n, min_val=0, max_val=100):
    """Tạo mảng 2 chiều (m x n) chứa các số nguyên ngẫu nhiên"""
    mang = []
    for _ in range(m):
        dong = [random.randint(min_val, max_val) for _ in range(n)]
        mang.append(dong)
    return mang

def xuat_mang(mang):
    """In mảng 2 chiều ra màn hình"""
    if not mang:
        print("Mảng rỗng.")
        return
    for dong in mang:
        print(dong)
    print("\n")

def xuat_dong_k(mang, k):
    """Xuất các phần tử thuộc dòng k"""
    if not mang or k < 0 or k >= len(mang):
        print(f"Dòng {k} không hợp lệ hoặc nằm ngoài phạm vi")
        return
    print(f"Các phần tử thuộc dòng {k}: {mang[k]}")

def xuat_cot_k(mang, k):
    """Xuất các phần tử thuộc cột k."""
    if not mang or not mang[0] or k < 0 or k >= len(mang[0]):
        print(f"Cột {k} không hợp lệ hoặc nằm ngoài phạm vi")
        return
    phan_tu_cot = [mang[i][k] for i in range(len(mang))]
    print(f"Các phần tử thuộc cột {k}: {phan_tu_cot}")

def tim_dong_tong_lon_nhat(mang):
    """Tìm dòng có tổng lớn nhất"""
    if not mang:
        print("Mảng rỗng, không thể tìm dòng có tổng lớn nhất")
        return -1, None

    tong_lon_nhat = -float('inf')
    chi_so_dong = -1

    for i, dong in enumerate(mang):
        tong_hien_tai = sum(dong)
        if tong_hien_tai > tong_lon_nhat:
            tong_lon_nhat = tong_hien_tai
            chi_so_dong = i

    print(f"Dòng có tổng lớn nhất là dòng {chi_so_dong} với tổng = {tong_lon_nhat}")
    return chi_so_dong, tong_lon_nhat

def tim_cot_tich_nho_nhat(mang):
    """Tìm cột có tích nhỏ nhất"""
    if not mang or not mang[0]:
        print("Mảng rỗng hoặc không hợp lệ, không thể tìm cột có tích nhỏ nhất")
        return -1, None

    so_cot = len(mang[0])
    tich_nho_nhat = float('inf')
    chi_so_cot = -1

    for j in range(so_cot):
        tich_hien_tai = 1
        for i in range(len(mang)):
            tich_hien_tai *= mang[i][j]

        if tich_hien_tai < tich_nho_nhat:
            tich_nho_nhat = tich_hien_tai
            chi_so_cot = j

    print(f"Cột có tích nhỏ nhất là cột {chi_so_cot} với tích = {tich_nho_nhat}")
    return chi_so_cot, tich_nho_nhat

def xuat_dong_chan_cot_le(mang):
    """Xuất ra các phần tử thuộc dòng chẵn và cột lẻ trong a"""
    print("Các phần tử thuộc dòng chẵn và cột lẻ:")
    phan_tu = []
    if not mang or not mang[0]:
        print("Mảng rỗng.")
        return phan_tu

    for i in range(len(mang)):
        if i % 2 == 0:  # Dòng chẵn
            for j in range(len(mang[0])):
                if j % 2 != 0:  # Cột lẻ
                    phan_tu.append(mang[i][j])
    print(phan_tu)
    return phan_tu

def tbc_phan_tu_chan_dong_le(mang):
    """Tính trung bình cộng các phần tử chẵn thuộc dòng lẻ của a"""
    tong = 0
    dem = 0
    if not mang:
        print("Mảng rỗng.")
        return 0

    for i in range(len(mang)):
        if i % 2 != 0:  # Dòng lẻ
            for phan_tu in mang[i]:
                if phan_tu % 2 == 0:  # Phần tử chẵn
                    tong += phan_tu
                    dem += 1

    if dem == 0:
        print("Không có phần tử chẵn nào thuộc dòng lẻ để tính trung bình cộng")
        return 0
    tbc = tong / dem
    print(f"Trung bình cộng các phần tử chẵn thuộc dòng lẻ: {tbc}")
    return tbc

def tbc_phan_tu_bien(mang):
    """Tính trung bình cộng các phần tử thuộc biên"""
    if not mang or not mang[0]:
        print("Mảng rỗng hoặc không hợp lệ")
        return 0

    m = len(mang)
    n = len(mang[0])
    tong = 0
    dem = 0

    if m == 1:
        tong = sum(mang[0])
        dem = n
    elif n == 1:
        for i in range(m):
            tong += mang[i][0]
        dem = m
    else:

        tong += sum(mang[0])
        tong += sum(mang[m-1])
        dem += n * 2


        for i in range(1, m - 1):
            tong += mang[i][0]
            tong += mang[i][n-1]
            dem += 2

    if dem == 0:
        print("Không có phần tử biên để tính trung bình cộng")
        return 0
    tbc = tong / dem
    print(f"Trung bình cộng các phần tử thuộc biên: {tbc}")
    return tbc

def tich_tbc_phan_tu_khong_bien(mang):
    """Tính trung bình tích các phần tử không thuộc biên (Geometric Mean)"""
    if not mang or not mang[0]:
        print("Mảng rỗng hoặc không hợp lệ")
        return 0

    m = len(mang)
    n = len(mang[0])
    phan_tu_khong_bien = []


    if m <= 2 or n <= 2:
        print("Không có phần tử không thuộc biên trong mảng này để tính trung bình tích")
        return 0

    for i in range(1, m - 1):
        for j in range(1, n - 1):
            phan_tu_khong_bien.append(mang[i][j])

    if not phan_tu_khong_bien:
        print("Không có phần tử không thuộc biên để tính trung bình tích")
        return 0

    tich_tong = 1
    for val in phan_tu_khong_bien:
        tich_tong *= val


    tbc_tich = tich_tong ** (1 / len(phan_tu_khong_bien))

    print(f"Tích các phần tử không thuộc biên: {tich_tong}")
    print(f"Số lượng phần tử không thuộc biên: {len(phan_tu_khong_bien)}")
    print(f"Trung bình tích (Geometric Mean) các phần tử không thuộc biên: {tbc_tich}")
    return tbc_tich


# --- Ví dụ sử dụng các hàm ---
so_dong = 5 # Số dòng
so_cot = 6 # Số cột

print(f"Tạo mảng {so_dong}x{so_cot} với các số nguyên ngẫu nhiên từ 0 đến 100:")
matran_cua_toi = tao_mang_ngau_nhien(so_dong, so_cot)
xuat_mang(matran_cua_toi)

# b. Xuất các phần tử thuộc dòng k
dong_k = 2
xuat_dong_k(matran_cua_toi, dong_k)

# c. Xuất các phần tử thuộc cột k
cot_k = 3
xuat_cot_k(matran_cua_toi, cot_k)

# d. Tìm dòng có tổng lớn nhất
tim_dong_tong_lon_nhat(matran_cua_toi)

# e. Tìm cột có tích nhỏ nhất
tim_cot_tich_nho_nhat(matran_cua_toi)

# f. Xuất ra các phần tử thuộc dòng chẵn và cột lẻ
xuat_dong_chan_cot_le(matran_cua_toi)

# g. Tính trung bình cộng các phần tử chẵn thuộc dòng lẻ
tbc_phan_tu_chan_dong_le(matran_cua_toi)

# h. Tính trung bình cộng các phần tử thuộc biên
tbc_phan_tu_bien(matran_cua_toi)

# i. Tính trung bình tích các phần tử không thuộc biên
tich_tbc_phan_tu_khong_bien(matran_cua_toi)

# Ví dụ với mảng nhỏ không có phần tử không thuộc biên
print("\n--- Ví dụ mảng nhỏ hơn (3x3) ---")
mang_nho = tao_mang_ngau_nhien(3, 3, 0, 10)
xuat_mang(mang_nho)
tich_tbc_phan_tu_khong_bien(mang_nho)

print("\n--- Ví dụ mảng 1x5 ---")
mang_mot_dong = tao_mang_ngau_nhien(1, 5, 0, 10)
xuat_mang(mang_mot_dong)
tbc_phan_tu_bien(mang_mot_dong)

print("\n--- Ví dụ mảng 5x1 ---")
mang_mot_cot = tao_mang_ngau_nhien(5, 1, 0, 10)
xuat_mang(mang_mot_cot)
tbc_phan_tu_bien(mang_mot_cot)
```

```python
"""Bài 15"""
class SinhVien:
    def __init__(self, ma_sinh_vien, ten, nam_sinh, diem_trung_binh):
        if not isinstance(ma_sinh_vien, str) or len(ma_sinh_vien) != 10:
            raise ValueError("Mã sinh viên phải là chuỗi có 10 ký tự")
        if not isinstance(ten, str) or len(ten) > 20:
            raise ValueError("Tên sinh viên phải là chuỗi tối đa 20 ký tự")
        if not isinstance(nam_sinh, int) or nam_sinh <= 0:
            raise ValueError("Năm sinh phải là số nguyên dương")
        if not isinstance(diem_trung_binh, (int, float)) or not (0 <= diem_trung_binh <= 10):
            raise ValueError("Điểm trung bình phải là số thực từ 0 đến 10")

        self.ma_sinh_vien = ma_sinh_vien
        self.ten = ten
        self.nam_sinh = nam_sinh
        self.diem_trung_binh = diem_trung_binh

    def __str__(self):
        return f"Mã SV: {self.ma_sinh_vien}, Tên: {self.ten}, Năm sinh: {self.nam_sinh}, ĐTB: {self.diem_trung_binh}"

def nhap_danh_sach_sinh_vien(n):
    danh_sach_sv = []
    for i in range(n):
        print(f"\nNhập thông tin cho sinh viên thứ {i+1}:")
        while True:
            try:
                ma_sv = input("Nhập mã sinh viên (10 ký tự): ")
                ten_sv = input("Nhập tên sinh viên (tối đa 20 ký tự): ")
                nam_sinh_sv = int(input("Nhập năm sinh: "))
                diem_tb_sv = float(input("Nhập điểm trung bình: "))
                sv = SinhVien(ma_sv, ten_sv, nam_sinh_sv, diem_tb_sv)
                danh_sach_sv.append(sv)
                break
            except ValueError as e:
                print(f"Lỗi nhập liệu: {e}. Vui lòng nhập lại")
    return danh_sach_sv

# Ví dụ sử dụng:
# Nhập số lượng sinh viên bạn muốn thêm
num_students = int(input("Nhập số lượng sinh viên bạn muốn thêm vào danh sách: "))

# Gọi hàm để nhập danh sách sinh viên
danh_sach_sinh_vien_moi = nhap_danh_sach_sinh_vien(num_students)

print("\n--- Danh sách sinh viên đã nhập ---")
if danh_sach_sinh_vien_moi:
    for sv in danh_sach_sinh_vien_moi:
        print(sv)
else:
    print("Không có sinh viên nào được nhập.")
```
